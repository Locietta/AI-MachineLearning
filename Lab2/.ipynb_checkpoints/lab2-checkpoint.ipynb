## 4.2.1 画出一个sin(x)图


```python
import matplotlib.pyplot as plt
import numpy as np

plt.xlabel('x Axis')
plt.ylabel('y Axis')

lx = np.linspace(-np.pi,np.pi,500)

plt.plot(lx,np.sin(lx))
plt.show()
```

## 4.2.2 选取一些点并加入高斯噪声

```python
def guass_noise(signal, n: int, SNR = 10): # 对指定信号输入signal, n和信噪比SNR产生高斯噪声
    noise = np.random.randn(n)
    noise = noise - np.mean(noise) #噪声均值为0
    signal_power = np.linalg.norm(signal)**2
    noise_variance = signal_power/np.power(10,(SNR/10))
    noise = (np.sqrt(noise_variance)/np.std(noise))*noise
    return noise

plt.xlabel('x Axis')
plt.ylabel('y Axis')

lx = np.linspace(-np.pi,np.pi,500)

plt.plot(lx,np.sin(lx), label = 'real sound')

n_sample_dot = 20
np.random.seed(1279)
x = np.linspace(-np.pi,np.pi,n_sample_dot)
y = np.sin(x)+guass_noise(np.sin(x),n_sample_dot, SNR = 25)
plt.scatter(x, y, color = '#1f1e33', label = 'samples with Guass noise')
plt.legend(loc = 'upper left')
```

## 4.2.3用sklearn的回归模型对噪声点进行拟合

分别采用3阶，9阶，20阶多项式回归进行拟合


```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

def poly_reg(X, Y, degree= 2):
    featurizer = PolynomialFeatures(degree = degree)
    poly = featurizer.fit_transform(X)
    reg = LinearRegression()
    reg.fit(poly, Y)
    return reg

plt.figure()
plt.subplot(1,3,1)
plt.xlabel('x Axis')
plt.ylabel('y Axis')

lx = np.linspace(-np.pi,np.pi,500)

plt.plot(lx,np.sin(lx), label = 'real sound')

n_sample_dot = 20
np.random.seed(1279)
x = np.linspace(-np.pi,np.pi,n_sample_dot)
y = np.sin(x)+guass_noise(np.sin(x),n_sample_dot, SNR = 25)

plt.scatter(x, y, color = '#1f1e33', label = 'samples with Guass noise')

plt.legend(loc = 'upper left')
```
